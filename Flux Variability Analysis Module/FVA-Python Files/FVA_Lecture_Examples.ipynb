# FVA Lecture Examples

## Maximum Growth FVA
Set Python environment and load the model

```python
import cobra
import pandas as pd
from cobrapy_bigg_client import client
pd.set_option('display.max_rows', 500)

model = client.download_model('e_coli_core', save=False) # Loading the BIGG model
```

Set the simulation conditions

```python
from cobra.flux_analysis import flux_variability_analysis
model.reactions.EX_glc__D_e.lower_bound = -0 # Must be set to zero if not a carbon source
model.reactions.EX_succ_e.lower_bound = -20
```

Run flux variability analysis

```python
fva=flux_variability_analysis(model, model.reactions,loopless = True, fraction_of_optimum=1.0)
fva.round(3)
```

The statistics of this flux variability analysis can be found as follows.

```python
fva.describe()
```

Find the difference (flux span) between the minimum and maximim or the flux span

```python
diff = fva.maximum - fva.minimum
difference = pd.DataFrame(diff)
difference.columns = ['Difference']

difference.round(6)
```

Identify the reactions that have a difference between the maximum and minimum flux values

```python
df = difference.round(6)
changes = df[df.loc[:]!=0].dropna() # remove all the rows with zeros
changes
```

Adding the flux differences to the FVA results in the fva dataframe

```python
fvaD = fva.join(difference.round(6))
fvaD
```

Plot box chart of the minimum and maximum values of the different reactions

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
df = np.transpose(fva);
df.plot.box(title="Flux Variability Analysis Box Plot", grid=False,figsize=(20, 10));
plot.xticks(rotation=90)
plot.show(block=True);
```

Plotting the difference on an Escher map

```python
import escher
from escher import Builder
builder = Builder(
    map_name='e_coli_core.Core metabolism',
    model_name='e_coli_core',
)
builder
```

```python
builder.reaction_data = difference.round(6)
```

## Objective Function Variablity

The objective function can be adjust so it can vary between it's optimum value and a certain percentage of that optimum value. This more accurately illustrates how the cell's growth rate can change during the growth process.
Perform flux variability analysis allowing a 5% variation in the objective function.

```python
fva=flux_variability_analysis(model, model.reactions,loopless = True, fraction_of_optimum=0.95)
diff = fva.maximum - fva.minimum
difference2 = pd.DataFrame(diff)
difference2.columns = ['Difference']
fvaD = fva.join(difference2.round(6))
fvaD
```

List the reactions that have changes in their flux

```python
df = difference2.round(6)
changes = df[df.loc[:]!=0].dropna() # remove all the rows with zeros
changes
```

Plot box chart of the minimum and maximum values of the different reactions. Note the large increase in reactions with changing flux values

```python
df = np.transpose(fva);
df.plot.box(title="Flux Variability Analysis Box Plot", grid=False,figsize=(20, 10));
plot.xticks(rotation=90)
plot.show(block=True);
```

Plot an Escher map of the flux differences

```python
import escher
from escher import Builder
builder2 = Builder(
    map_name='e_coli_core.Core metabolism',
    model_name='e_coli_core',
)
builder2
```

```python
builder2.reaction_data = difference2.round(6)
```

## FVA in Summary Methods

Flux variability analysis can also be embedded in calls to summary methods. As an example, the expected variability in substrate consumption and product formation can be found by

```python
model.optimize()
model.summary(fva=0.95)
```

Note that the secretion values are the negative of the normal format for exchange reactions which says that poitive fluxes are secreted while negative fluxes refer to uptake.

Variability in metabolite mass balances can also be checked with flux variability analysis.

```python
model.metabolites.pyr_c.summary(fva=0.95)
```

Look at the variability for a given reaction.

```python
model.reactions.ACALD.summary(fva=0.95)
```

## Removing Loops
The standard FVA may contain loops (futile cycles), i.e. high absolute flux values that only can be high if they are allowed to participate in loops (a mathematical artifact that cannot happen in vivo). Use the loopless argument to avoid such loops. Below, we can see that FRD7 and SUCDi reactions can participate in loops but that this is avoided when using the looplesss FVA. Care must be taken using the loopless feature; there are times that the loopless feature will not allow the FVA algorithm to converge! 

```python
model.reactions.EX_glc__D_e.lower_bound = -20 # Must be set to zero if not a carbon source
model.reactions.EX_succ_e.lower_bound = -0
model.objective = 'BIOMASS_Ecoli_core_w_GAM'
result = flux_variability_analysis(model)
result[result.maximum > 500]
```

Looking specifically at the two loop reactions

```python
loop_reactions = [model.reactions.SUCDi,model.reactions.FRD7]
flux_variability_analysis(model, reaction_list=loop_reactions, loopless=False)
```

Removing the loops by setting "loopless=True"

```python
flux_variability_analysis(model, reaction_list=loop_reactions, loopless=True)
```

## Flux Variability Analysis and Parsimoneous Flux Balance Analysis

Within the FVA package is the ability to include a pfba_factor. This implies that we are not interested in solutions that have very low total absolute flux (e.g. 1.1 which implies no more than 10% over the minimum total flux).

```python
flux_variability_analysis(model, fraction_of_optimum=1.0, pfba_factor=1.1, loopless=True)
```
