# Flux Variability Analysis Lab

### Problem 1. ###
Using the *E.coli* core model find the flux variability in the fluxes moving in and out of the cell when the carbon source, glucose, has a lower bound of -20 mmol/gDW-h in an anaerobic environment where the optimum value of the objective function to vary by 10%. Also, create a box plot showing the variability of the fluxes in the model under these conditions.

```python
import cobra.test
from cobra.flux_analysis import flux_variability_analysis
import escher
from escher import Builder
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
pd.set_option('display.max_rows', 1000)
model = cobra.test.create_test_model("textbook")
```

Set the simulation conditions

```python
model.reactions.EX_glc__D_e.lower_bound = -20 
model.reactions.EX_o2_e.lower_bound = -0
```

Optimizing the model

```python
solution = model.optimize()
model.summary(fva=0.90)
```

Performing flux variability on the model and the creating a box plot showing the ranges of the different fluxes.

```python
fva=flux_variability_analysis(model, model.reactions,loopless = True, fraction_of_optimum=0.9)

bp_df = np.transpose(fva);
bp_df.plot.box(title="Flux Variability Analysis Box Plot", grid=False,figsize=(20, 10));
plot.xticks(rotation=90)
plot.show(block=True);
```

Plotting the fluxes on an Escher map

```python
builder = Builder(
    map_name= 'e_coli_core.Core metabolism',
    model_name='e_coli_core',
)
builder
```

```python
builder.reaction_data = solution.fluxes
```

### Problem 2 ### 
Using the *E.coli* core model use flux variability analysis to find the reactions that have flux variability and plot them on a box plot, then plot the flux difference on an Escher map. Assume a lower limit of -20 mmol/gDW-h for glucose as the carbon source and assume a lower limit of -0 mmol/gDW-h uptake for oxygen, and the biomass function is the objective function (90% of optimum).

Set Python environment and load the model

```python
model2 = cobra.test.create_test_model("textbook")
```

```python
model2.reactions.EX_glc__D_e.lower_bound = -20 
model2.reactions.EX_o2_e.lower_bound = -0
```

Run flux variability analysis

```python
%%time
fva=flux_variability_analysis(model2, model.reactions,loopless = True, fraction_of_optimum=0.9)
```

```python
diff = fva.maximum - fva.minimum
difference = pd.DataFrame(diff)
difference.columns = ['Difference']
fvaD = fva.join(difference.round(6))
fvaD
```

```python
df = np.transpose(fva);
df.plot.box(title="Flux Variability Analysis Box Plot", grid=False,figsize=(20, 10));
plot.xticks(rotation=90)
plot.show(block=True);
```

List the reactions that have changes in their flux

```python
df = difference.round(7)
changes = df[df.loc[:]!=0].dropna() # remove all the rows with zeros
changes
```

Plot flux variability on Escher map

```python
builder2 = Builder(
    map_name= 'e_coli_core.Core metabolism',
    model_name='e_coli_core',
)
builder2
```

```python
builder2.reaction_data = difference.round(6)
```
