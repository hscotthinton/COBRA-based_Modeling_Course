# Essential Genes/Reactions: Simulating Single Deletions in iJO1366

```python
import pandas as pd
from pandas import DataFrame
import cobra.test
from cobra.flux_analysis import (
    single_gene_deletion, single_reaction_deletion, double_gene_deletion,
    double_reaction_deletion)
pd.set_option('display.max_rows', 1000)

from cobrapy_bigg_client import client
model_orig = client.download_model('iJO1366', save=False) # Download model from the BIGG database
```

## Essential Genes

**An essential gene is a gene that when knocked out prevents cell growth!** 

First let's review how many genes and reactions are included in the iJO1366 model.

```python
print('The number of genes in the iJO1366 model = ',len(model_orig.genes))
print('The number of reactions in the iJO1366 model = ',len(model_orig.reactions))
print('The number of metabolites in the iJO1366 model = ',len(model_orig.metabolites))
```

Perform all single gene deletions on a model

```python
model = model_orig.copy()
deletion_results = single_gene_deletion(model)
deletion_results[0:20]
```

The essential genes are

```python
gene_deletion_results = deletion_results[deletion_results['growth'] == 0] # Remove genes with growth
print('Number of essential genes =',len(gene_deletion_results))
gene_deletion_results
```

Let's create a histogram plot of the growth-rates. The ones with zero growth-rate are teh essential reactions.

```python
deletion_results.plot.hist()
```

Let's plot the values of the growth-rate starting from zero to the largest value.

```python
deletion_results = deletion_results.sort_values('growth') # Sorting the growth-rate values
ax = deletion_results.plot.bar().axes
ax.figure.set_size_inches(16, 9)
```

Essential genes are genes that if knocked out will inhibit growth. The essential genes can be found by

```python
essentialGenes = deletion_results.loc[deletion_results['growth'] < 0.00001] # Low growth-rate determines essential genes
essentialGenes = essentialGenes.reset_index() # Renumber the rows
essentialGenes.round(4)
```

For the iJO1366 *E. coli* model under default conditions there are 207 essential genes. Note that the essential genes will change with the inital conditions of the model. A different carbon source, nitrogen source, oxygen level, etc. will each have different essential genes.

## Essential Reactions

An essential reaction is a reactions that when knocked out prevents cell growth! 

As with genes, this can also be done for reactions

```python
SingleReactionDeletions = single_reaction_deletion(model, model.reactions)
SingleReactionDeletions = SingleReactionDeletions.sort_values('growth') # Sorting the growth-rate values
SingleReactionDeletions = SingleReactionDeletions.reset_index() # Renumbering the rows
SingleReactionDeletions.round(4)[0:100] 
```

List the essential reactions

```python
reaction_deletion_results = SingleReactionDeletions[SingleReactionDeletions['growth'] == 0] 
print('Number of essential reactions =',len(reaction_deletion_results))
reaction_deletion_results
```

Creating a histogram of the reaction's growth-rates.

```python
SingleReactionDeletions.growth.plot.hist()
```

Let's plot the values of the reaction growth-rates starting from zero to the largest value.

```python
ax = SingleReactionDeletions.growth.plot.bar().axes
ax.figure.set_size_inches(16, 9)
```

The essential reactions can be found by

```python
essentialReactions = SingleReactionDeletions.loc[SingleReactionDeletions['growth'] < 0.001]
essentialReactions = essentialReactions.reset_index() # Renumber the rows
essentialReactions.round(4)
```

Note their are 287 essential reactions for the default model conditions.
