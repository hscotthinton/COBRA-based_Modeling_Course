# Robustness Analysis: *E.coli* core model

### Load appropriate Python and COBRApy packages

```python
import cobra.test
import numpy as np
import pandas as pd
from cobrapy_bigg_client import client
import matplotlib.pyplot as plt
pd.set_option('display.max_rows', 500)
```

### Load model and set constraints

```python
model_orig = client.download_model('e_coli_core', save=False) # Loading the model to the simulation
model_orig.reactions.EX_o2_e.lower_bound = -20
model_orig.reactions.EX_glc__D_e.lower_bound = -10
model = model_orig.copy()
```

## Robustness Analysis using the "production envelopes" COBRApy methodÂ¶

For more info see https://cobrapy.readthedocs.io/en/latest/phenotype_phase_plane.html

### Robustness analysis for oxygen

```python
from cobra.flux_analysis import production_envelope

model = model_orig.copy()
RA_o2 = production_envelope(
    model, ["EX_o2_e"], objective="BIOMASS_Ecoli_core_w_GAM", carbon_sources="EX_glc__D_e", points=100)
RA_o2.round(5)
```

```python
RA_o2.plot(kind='line', x='EX_o2_e', y='flux_maximum',figsize=(10,7), fontsize = 12, grid = True,
          ylabel = "Objective", title = 'Robustness Analysis of Oxygen')
```
