## iJN678 Subsystems

List the subsystem to determine the name of the oxidative phosphorylation subsystem.

```python
reaction_subsystem = [r.subsystem for r in model.reactions]
subsystem = pd.DataFrame(reaction_subsystem, columns = ['iJN678 Subsystems'])
subsystem = subsystem.drop_duplicates()
subsystem = subsystem.reset_index(drop=True)
subsystem
```

Create a table of all the reactions in the oxidative phosphorylation subsystem

```python
reaction_names = [r.name for r in model.reactions]
reaction_ids = [r.id for r in model.reactions]
reaction_formula = [r.reaction for r in model.reactions]
reaction_subsystem = [r.subsystem for r in model.reactions]
reactionList = {'Reaction ID': reaction_ids,
                'Reaction Name': reaction_names,
                'Reaction Formula': reaction_formula,
                'Reaction Subsystem': reaction_subsystem,
               }

df = pd.DataFrame(reactionList, columns= ['Reaction ID','Reaction Name','Reaction Formula','Reaction Subsystem'])
# Select the reactions associated with the desired subsystem
SSindex = []
for i in range(len(reaction_subsystem)):
    if reaction_subsystem[i] == 'Oxidative phosphorylation':
        SSindex.append(i)
df_ss = df.loc[SSindex]
df_ss = df_ss.sort_values(by='Reaction ID')
df_ss = df_ss.reset_index(drop=True)
df_ss
```

## Setting up the Heterotrophic Environment

Setting the environment for heterotrophic operations [1]

```python
model = model_orig.copy()
# Set Biomass Function
model.objective = 'BIOMASS_Ec_SynHetero'
model.reactions.BIOMASS_Ec_SynAuto.bounds = [-0,0]
model.reactions.BIOMASS_Ec_SynMixo.bounds = [-0,0]

#model.reactions.EX_glc__D_e.bounds=[-0.85,0]
model.reactions.EX_glc__D_e.bounds=[-3.7,0]
model.reactions.HCO3E.bounds=[-0,0]

# Turn off ferrocytochrome c6 based reactions
model.reactions.CYO1bpp_syn.bounds = [-0,0]
model.reactions.CBFC2pp.bounds = [-0,0]

# Turn off photosynthetic reactions
model.reactions.EX_photon_e.bounds=[-0,0]
model.reactions.PSII.bounds=[-0,0]
model.reactions.CBFCu.bounds=[-0,0]
model.reactions.PSI.bounds=[-0,0]
model.reactions.CYO1b2_syn.bounds=[-0,0]
model.reactions.ATPSu.bounds=[-0,0]
model.reactions.CBFC2.bounds=[-0,0]
model.reactions.CYO1b_syn.bounds=[-0,0]
model.reactions.NDH1_1u.bounds=[-0,0]
model.reactions.NDH1_2u.bounds=[-0,0]
model.reactions.NDH1_3u.bounds=[-0,0]
model.reactions.NDH2_syn.bounds=[-0,0]
model.reactions.FNOR.bounds=[-0,0]

model_h = model.copy() # Save heterotropic model
```

```python
model = model_h.copy()
RA_o2 = production_envelope(
    model, ["EX_o2_e"], objective="BIOMASS_Ec_SynHetero", carbon_sources="EX_glc__D_e", points=100)
RA_o2.round(5)
```

### Oxidative Phosphorylation

The aerobic respiration (oxidative phosphoylation) pathways illustrated in the Nogales paper [1] is shown below.



The main electron transport pathway begins with NADH (nadh_c) entering the nadh_c flux at the NAD(P)H dehydrogenase NDH2 (NDH1_2p) which passes through the plastoquinone (PQ) pool to the cytochrome b6/f complex CYTBF (CBFCpp). From there the plastocynanin carriers move the electrons to the cytochrome c oxidase CYO (CYO1b2pp_syn). The results of this movement through the electron transport chain is that protons have been transferred to the periplasmic space where they create a proton-motive force that will force the electrons from the periplasm to pass through the ATP synthase (ATPS4rpp_1) creating cytoplasmic ATP (atp_c).

Let's begin by looking at the flux associated with the reactions in the electron transport chain and oxidative phosphorylation subsystem as the glucose uptake is increased from 0 to 0.85.

```python
%matplotlib inline
#h_reactions = list(df_ss['Reaction ID'])
h_reactions = ['NDH1_2p','SUCDpp_syn','CBFCpp','CYO1b2pp_syn','ATPS4rpp_1','NDH1_1p',
                       'CBFC2pp','CYO1bpp_syn','CYTBDpp_1','NADH5'] # Reactions in flux flow order

tsolution = []
x=0
for i in range(100):
    model.reactions.EX_glc__D_e.bounds = [-x,-x] # Adjust the photon flux to go from 0 to -100
    tsolution.append(model.optimize()) # For each photon flux value do FBA
    x +=0.85/100

df = pd.DataFrame()
for reaction in h_reactions:
    f=[]
    for i in range(100):
        f.append(tsolution[i].fluxes[reaction]) # Create a list of fluxes for a given reaction
    df_temp = pd.DataFrame(f) # Convert the list to a Pandas Dataframe
    df[reaction] = df_temp # Add the new dataframe to the dataframe that will hold all the fluxes
df.round(4)
```

Let's plot the results

```python
df.plot(subplots=True, layout=(4,5),figsize=(15,11));
```

From these results we can see that the standard flow though the electron transport chain includes the following periplasmic membrane reactions, NDH1_2p, CBFCpp, CYO1b2pp_syn, and ATPS4rpp_1.

In heterotrophic growth conditions, glucose was funneled mainly through the oxidative pentose phosphate (OPP) pathway, followed by the NAD-dependent glyceraldehyde-3-phosphate dehydrogenase (GAPD), which drives the carbon flux to the incomplete TCA cycle. The high flux through the oxidative pentose phosphate pathway, and the flux across GAPD, as well as the significant flux across the TCA cycle (mainly the NADPH- dependent isocitrate dehydrogenase) provides reducing power in the form of NADPH. In addition, the TCA cycle supplied important biosynthetic precursors together with carbon skeletons for nitrogen fixation in the form of 2-oxoglutarate. Succinate was generated through the GABA shunt, while the succinate semialdehyde dehydrogenase (SSALy) further increased the NADPH levels. Finally, the reducing power was oxidized by the NADPH and SDH dehydrogenase complexes, contributing to energy production under heterotrophic conditions.[1]
