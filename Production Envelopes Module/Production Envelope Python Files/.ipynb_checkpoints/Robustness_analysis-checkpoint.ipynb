# Robustness Analysis

Loading the appropriate Python and COBRApy packages

```python
import cobra.test
import numpy as np
import pandas as pd
from cobrapy_bigg_client import client
import matplotlib.pyplot as plt
pd.set_option('display.max_rows', 500)
```

## Robustness Analysis Function

Define Robustness Analysis [robustnessAnalysis()] function (method)

```python
def robustnessAnalysis(model, ra_reaction, l_bound=-20, u_bound=0, npts=100):
    '''Robustness Analysis: to calculate robustness analysis the upper and lower bounds need to be equal'''
    model.reactions.get_by_id(ra_reaction).lower_bound = l_bound
    model.reactions.get_by_id(ra_reaction).upper_bound = l_bound
    BM_objective = []
    reaction_lb = []
    increment = (u_bound - l_bound)/npts
    for i in range(npts):
        BM_objective.append(model.slim_optimize())
        model.reactions.get_by_id(ra_reaction).upper_bound += increment
        model.reactions.get_by_id(ra_reaction).lower_bound = model.reactions.get_by_id(ra_reaction).upper_bound
        reaction_lb.append(model.reactions.get_by_id(ra_reaction).lower_bound)
    plt.plot(reaction_lb, BM_objective)
    plt.show()
```

## Robustness Analysis using "robustnessAnalysis" Function

Set model environmental conditions calculate robustness analysis for EX_glc__D_e.

```python
model = client.download_model('e_coli_core', save=False) # Loading the model to the simulation
model_orig = model.copy()

model.reactions.EX_o2_e.lower_bound = -20
model.reactions.EX_glc__D_e.lower_bound = -10
robustnessAnalysis(model, 'EX_glc__D_e', -30, 0)  
```

Calculate robustness analysis for EX_o2_e.

```python
model2 = model_orig.copy()
model2.reactions.EX_o2_e.lower_bound = -0
model2.reactions.EX_glc__D_e.lower_bound = -10
robustnessAnalysis(model2, 'EX_o2_e', -60, 0)  
```

Calculate robustness analysis for EX_etoh_e.

```python
model3 = model_orig.copy()
model3.reactions.EX_o2_e.lower_bound = 0
model3.reactions.EX_glc__D_e.lower_bound = -10
robustnessAnalysis(model2, 'EX_etoh_e', -10, 20) 
```

## Robustness Analysis Using "production_envelope" COBRApy Function¶

```python
model3.summary()
```

Robustness analysis for oxygen

```python
from cobra.flux_analysis import production_envelope

model4 = model_orig.copy()
RA_o2 = production_envelope(
    model4, ["EX_o2_e"], objective="BIOMASS_Ecoli_core_w_GAM", carbon_sources="EX_glc__D_e", points=100)
RA_o2.round(5)
```

```python
RA_o2.plot(kind='line', x='EX_o2_e', y='flux_maximum',figsize=(10,7))
```

Robustness analysis for glucose

```python
model4.reactions.EX_o2_e.lower_bound = -20
model4.reactions.EX_glc__D_e.lower_bound = -30
RA_glc = production_envelope(
    model4, ["EX_glc__D_e"], objective="BIOMASS_Ecoli_core_w_GAM", carbon_sources="EX_glc__D_e", points=100)
RA_glc.round(5)
```

```python
RA_glc.plot(kind='line', x='EX_glc__D_e', y='flux_maximum',figsize=(10,7))
```
