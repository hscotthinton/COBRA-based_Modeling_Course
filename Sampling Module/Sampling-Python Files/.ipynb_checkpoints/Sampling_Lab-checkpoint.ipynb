Setting initial conditions

```python
import cobra.test
from cobra.test import create_test_model
import numpy as np
import pandas as pd
from cobra.sampling import sample
import matplotlib.pyplot as plt
pd.set_option('display.max_rows', 500)
```

## Problem 1.
What is the difference in the sampling histograms between the default conditions and a condition where the bounds on the PGI reaction is limited from -5 to +5? Assume the oxygen uptake is limited to -20 mmol/gDW-hr and the glucose uptake is limited to -20 mmol/gDW-hr.

### Default conditions simulation

```python
model = create_test_model("textbook")
model.reactions.EX_o2_e.lower_bound = -20
model.reactions.EX_glc__D_e.lower_bound = -20
```

```python
model.summary()
```

Default simulation

```python
default = sample(model, 10000, processes=8) #Set this to 1000 for faster, less accurate performance
default
```

## Constrained condition simulation

```python
model2 = create_test_model("textbook")
model2.reactions.EX_o2_e.lower_bound = -20
model2.reactions.EX_glc__D_e.lower_bound = -20
model2.reactions.PGI.bounds = [-1,1]
model2.reactions.PGI.bounds
```

```python
model2.summary()
```

```python
constrained = sample(model2, 10000, processes=8) #Set this to 1000 for faster, less accurate performance
constrained 
```

Now let's print out a collection of subplots comparing all the reactions in the previous list

```python
reactions = ['PGI','PFK','FBP','FBA','TPI','GAPD','PGK','PGM','ENO','PYK','PPS','PDH']
fig, axes = plt.subplots(nrows=3, ncols=4)
j=0
for i in range(len(reactions)):
    default_data=default[reactions[i]]
    constrained_data=constrained[reactions[i]]
    test = reactions[i]
    temp = pd.DataFrame({reactions[i]+ ': Default data':default_data, "Constrained Data":constrained_data})
    temp.plot.hist(bins=250, histtype = 'step',ax=axes[j%3,i%4],figsize=(15,15),density=True)
    if (i != 0 and (i+1)%4 == 0):
        j += 1
```
