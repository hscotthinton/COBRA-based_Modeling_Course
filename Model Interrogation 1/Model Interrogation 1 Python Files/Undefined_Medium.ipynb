# Undefined vs. Minimal Media

```python
from pandas import DataFrame
import pandas as pd
from cobra.sampling import sample
import matplotlib.pyplot as plt
from cobrapy_bigg_client import client
pd.set_option('display.max_rows', 500)

model_orig = client.download_model('e_coli_core', save=False) # Download model from the BIGG database
nsamples = 5000
```

## Minimal Medium

Let's begin by looking at the open exchanges needed to represent minimal media. For small models "model.medium" can be used but with larger models it becomes more awkward to deal with the large amount of data. Remember that the "model.medium" command shows the lower bound numbers as positive instead of the negative values that are normally used for uptake.

```python
model = model_orig.copy()
#model.reactions.EX_o2_e.bounds = [0,0] # Creating anaerobic conditions
minimal_exchanges = pd.DataFrame(list(model.medium.items()), columns = ['Exchange Reactions','| Lower Bound |'])
minimal_exchanges
```

Looking at the model summary for the flux flowing into the cell

```python
model.summary()
```

Note that there are only five open exchanges that will support uptake of metabolites. Also note that the growth-rate is 0.211.

Calculating randomized samples for the minimal media model.

```python
minimal = sample(model, nsamples, processes=8) #Set this to 1000 for faster, less accurate performance
minimal
```

## Undefined Medium

Allow all exchange reactions to uptake. For small models "model.medium" can be used but with larger models it becomes more awkward to deal with the large amount of data. Remember that the "model.medium" command shows the lower bound numbers as positive instead of the negative values that are normally used for uptake.

```python
model2 = model_orig.copy()
for reaction in model2.exchanges:
    if reaction.lower_bound == 0:
        reaction.lower_bound = -5
#model2.reactions.EX_o2_e.bounds = [0,0] # Creating anaerobic conditions
undefined_exchanges = pd.DataFrame(list(model2.medium.items()), columns = ['Exchange Reactions','| Lower Bound |'])
undefined_exchanges
```

Looking at the model summary for the flux flowing into the cell

```python
model2.summary()
```

Note the significant increase in open exchanges to 12 compared to 5 in the minimal case. Also note that the growth-rate is 0.957, over 4.5 times the growth-rate of the minimal media. 

Calculating randomized samples for the undefined media model.

```python
undefined = sample(model2, nsamples, processes=8) #Set this to 1000 for faster, less accurate performance
undefined
```

```python
reactions = ['PGI','PFK','FBP','FBA','TPI','GAPD','PGK','PGM','ENO','PYK','BIOMASS_Ecoli_core_w_GAM','ATPM']
fig, axes = plt.subplots(nrows=3, ncols=4)
j=0
for i in range(len(reactions)):
    minimal_data=minimal[reactions[i]]
    undefined_data=undefined[reactions[i]]
    test = reactions[i]
    temp = pd.DataFrame({reactions[i]+ ': minimal data':minimal_data, "undefined Data":undefined_data})
    temp.plot.hist(bins=250, histtype = 'step',ax=axes[j%3,i%4],figsize=(15,15),density=True)
    if (i != 0 and (i+1)%4 == 0):
        j += 1
```

From these charts we can see that the solution space has changed due to the additional open exchanges creating different flux distributions. Since there are more carbon sources that can provide flux to the cell we should expect to see the flux distributions shifted over a larger area.
